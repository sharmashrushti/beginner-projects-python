```python
# Mad Libs Generator in Python

print("Welcome to the Mad Libs Generator!\n")

# Collecting inputs from the user
name = input("Enter a name: ")
place = input("Enter a place: ")
adjective = input("Enter an adjective: ")
animal = input("Enter an animal: ")
verb = input("Enter a verb (past tense): ")
food = input("Enter a food item: ")

# Story template
story = f"""
One day, {name} went to {place}.
It was a very {adjective} day.

Suddenly, a {animal} appeared and started to {verb} everywhere!
Everyone was shocked.

To calm things down, {name} offered it some {food}.
Surprisingly, the {animal} became friendly and the day became unforgettable!
"""

# Displaying the final story
print("\n----- Your Mad Libs Story -----")
print(story)
```
