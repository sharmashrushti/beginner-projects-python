```python
# Unit Converter in Python

def length_converter(value, unit):
    if unit == "m_to_km":
        return value / 1000
    elif unit == "km_to_m":
        return value * 1000
    elif unit == "cm_to_m":
        return value / 100
    elif unit == "m_to_cm":
        return value * 100
    else:
        return None


def weight_converter(value, unit):
    if unit == "kg_to_g":
        return value * 1000
    elif unit == "g_to_kg":
        return value / 1000
    elif unit == "kg_to_lb":
        return value * 2.20462
    elif unit == "lb_to_kg":
        return value / 2.20462
    else:
        return None


def temperature_converter(value, unit):
    if unit == "c_to_f":
        return (value * 9/5) + 32
    elif unit == "f_to_c":
        return (value - 32) * 5/9
    elif unit == "c_to_k":
        return value + 273.15
    elif unit == "k_to_c":
        return value - 273.15
    else:
        return None


while True:
    print("\n===== UNIT CONVERTER =====")
    print("1. Length")
    print("2. Weight")
    print("3. Temperature")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        value = float(input("Enter value: "))
        print("Options: m_to_km, km_to_m, cm_to_m, m_to_cm")
        unit = input("Enter conversion type: ")
        result = length_converter(value, unit)
        print("Result:", result)

    elif choice == "2":
        value = float(input("Enter value: "))
        print("Options: kg_to_g, g_to_kg, kg_to_lb, lb_to_kg")
        unit = input("Enter conversion type: ")
        result = weight_converter(value, unit)
        print("Result:", result)

    elif choice == "3":
        value = float(input("Enter value: "))
        print("Options: c_to_f, f_to_c, c_to_k, k_to_c")
        unit = input("Enter conversion type: ")
        result = temperature_converter(value, unit)
        print("Result:", result)

    elif choice == "4":
        print("Exiting...")
        break

    else:
        print("Invalid choice. Try again.")
```
